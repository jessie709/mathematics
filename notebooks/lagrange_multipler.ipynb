假设有方程 $x^2y=3$,其函数图如下, 求原点到该函数最近的距离.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x1 = np.linspace(-3, 0, endpoint=False)
x2 = -x1
y1 = 3 / np.square(x1)
y2 = 3 / np.square(x2)

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.linspace(-3, 0, 300, endpoint=False)
y1 = np.linspace(0, 100, 300)
x1, y1 = np.meshgrid(x1, y1)
z1 = x1**2 * y1
x2 = -1 * np.linspace(-3, 0, 300, endpoint=False)
y2 = np.linspace(0, 100, 300)
x2, y2 = np.meshgrid(x2, y2)
z2 = x2**2 * y2

ax.plot_surface(x1, y1, z1)
ax.plot_surface(x2, y2, z2)

plt.show()
```

```python
plt.contour(x1, y1, z1)
plt.contour(x2, y2, z2)
```
