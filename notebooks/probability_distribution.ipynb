# 概率分布（Probability Distribution）

## Probability mass function

概率质量函数（probability mass function）,通常使用 $P(x)$ 来表示，作为 概率质量函数 必须满足如下条件：

$$
\begin{cases}
    P(x) & \geq 0 & \quad \text{for all } x \\
    \Sigma_x P(x) &= 1
\end{cases}
$$

概率质量函数与概率的关系如下：

$$
Pr(X=x) = P(x)
$$

## Probability density function

概率密度函数（probability density function），通常使用 $\rho(x)$ 来表示，作为 概率密度函数 必须满足如下条件：

$$
\begin{cases}
    \rho(x) & \geq 0 & \quad \text{for all } x \\
    \int \rho(x)\, \mathrm{d}x &= 1
\end{cases}
$$

有了概率密度函数，我们就可以计算概率了，概率和概率密度函数的关系如下

$$
Pr(X \in (x, x+dx)) = \int \rho(x)\, \mathrm{d}x
$$

**Example:**

假设 $\rho(x)=3x^2 \, x \in [0,1]$，问 $x \in [\frac{1}{2}, 1]$ 的概率
> 这里必须注意 $x \in [0,1]$ 是非常关键的，如果没有这个约束，$\rho(x)=3x^2$ 函数不能作为概率密度函数的，因为必须满足 $\int \rho(x)\, \mathrm{d}x = 1$ 这个条件，对于 $\rho(x)=3x^2$ 其积分为
>
> $
\begin{align*}
\int \rho(x) &= \int 3x^2 \mathrm{d}x \\
          &= 3 \int x^2 \mathrm{d}x \\
          &= 3 \frac{x^3}{3} \\
          &= x^3
\end{align*}
$
>
> 只有 $x \in [0,1]$ 才能保证积分结果等于1 $1^3 - 0^3 = 1$

求解：

$
\begin{align*}
\int_{\frac{1}{2}}^{1} \rho(x) &= \bigg [ x^3 \bigg ]_{\frac{1}{2}}^{1} \\
                     &= 1^3 - (\frac{1}{2})^3 \\
                     &= 1 - \frac{1}{8} \\
                     &= \frac{7}{8}
\end{align*}
$

答：$x \in [\frac{1}{2}, 1]$ 的概率是 $\frac{7}{8}$
